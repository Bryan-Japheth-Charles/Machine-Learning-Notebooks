# Confusion Matrix

```python
uploaded = files.upload()
```

```python
df.head()
```

```python
df.drop(columns = ['city'], inplace = True)
```

```python
df.isnull().sum()
```

```python
si = SimpleImputer(strategy = 'mean')
```

```python
df['cgpa'] = si.fit_transform(df[['cgpa']])
df['iq'] = si.fit_transform(df[['iq']])
```

```python
x = df.iloc[:,:-1]
y = df['placement']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2)
```

```python
lr3 = LogisticRegression()
```

```python
lr3.fit(x_train, y_train)
```

```python
lr3.score(x_test, y_test)*100
```

```python
from sklearn.metrics import confusion_matrix, f1_score, precision_score, recall_score
```

```python
cf = confusion_matrix(y_test,lr3.predict(x_test))
cf
```

```python
sns.heatmap(cf, annot = True)
plt.show()
```

```python
precision_score(y_test,lr3.predict(x_test))*100
```

```python
f1_score(y_test,lr3.predict(x_test))*100
```

```python
recall_score(y_test,lr3.predict(x_test))*100
```
